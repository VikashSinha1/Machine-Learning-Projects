```python
df=df.drop(['Over18','StandardHours','EmployeeCount','EmployeeNumber'],axis=1)
```

```python
df
```

now lets convert the categorical values into numeric values

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
list1=['Attrition','BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','OverTime']
for val in list1:
    df[val]=le.fit_transform(df[val])
        
```

now lets split the data for model building purpose

```python
x=df.drop('Attrition',axis=1)
y=df['Attrition']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.ensemble import RandomForestClassifier
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=100)
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
lr.score(x_train,y_train)
predlr=lr.predict(x_test)
print(accuracy_score(y_test,predlr))
print(classification_report(y_test,predlr))
print(confusion_matrix(y_test,predlr))
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gb=GaussianNB()
gb.fit(x_train,y_train)
gb.score(x_train,y_train)
predgb=gb.predict(x_test)
print(accuracy_score(y_test,predgb))
print(classification_report(y_test,predgb))
print(confusion_matrix(y_test,predgb))
```

we can see that precision and recall is good for 0 but not good for 1

```python
sns.countplot(df['Attrition'])
```

our output variable is imbalanced. lets make it  balanced. 

```python
from imblearn.over_sampling import SMOTE
```

```python
smt=SMOTE()
trainx,trainy=smt.fit_resample(x,y)
```

```python
trainx.value_counts()
```

```python
trainy.value_counts()
```

```python
sns.countplot(trainy)
```

```python
from sklearn.model_selection import train_test_split
```

```python

train_x,test_x,train_y,test_y=train_test_split(trainx,trainy,test_size=0.25,random_state=100)
```

```python
gb=GaussianNB()
gb.fit(train_x,train_y)
gb.score(train_x,train_y)
predgb=gb.predict(test_x)
print(accuracy_score(test_y,predgb))
print(classification_report(test_y,predgb))
print(confusion_matrix(test_y,predgb))
```

as we can see the precision and recall has beed improved for both 0 and 1, lets try some other algorithms

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
svc.fit(train_x,train_y)
svc.score(train_x,train_y)
predsvc=svc.predict(test_x)
print(accuracy_score(test_y,predsvc))
print(classification_report(test_y,predsvc))
print(confusion_matrix(test_y,predsvc))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
dt.fit(train_x,train_y)
dt.score(train_x,train_y)
preddt=dt.predict(test_x)
print(accuracy_score(test_y,preddt))
print(classification_report(test_y,preddt))
print(confusion_matrix(test_y,preddt))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=20)
rf.fit(train_x,train_y)
rf.score(train_x,train_y)
predrf=rf.predict(test_x)
print(accuracy_score(test_y,predrf))
print(classification_report(test_y,predrf))
print(confusion_matrix(test_y,predrf))
```

lets check the best parameters using gridsearchcv

```python
from sklearn.model_selection import GridSearchCV
import numpy as np
```

```python
n_estimators=[int(x) for x in np.linspace(start=10,stop=80,num=10)]
max_features=['auto','sqrt']
max_depth=[2,4]
min_samples_split=[2,5]
min_samples_leaf=[1,2]
bootstrap=[True,False]
```

```python
grid_param={"n_estimators": n_estimators,
           'max_features': max_features,
           'max_depth': max_depth,
           'min_samples_split': min_samples_split,
           'min_samples_leaf': min_samples_leaf,
            'bootstrap': bootstrap}
```

```python

print(grid_param)
```

```python
rf=RandomForestClassifier()
```

```python
rf_grid=GridSearchCV(estimator=rf,param_grid=grid_param,cv=3,verbose=2,n_jobs=4)
```

```python
rf_grid.fit(train_x,train_y)
```

```python
rf_grid.best_params_
```

```python
print(rf_grid.score(train_x,train_y))
print(rf_grid.score(test_x,test_y))
```
