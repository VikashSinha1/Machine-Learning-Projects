```python
def calc_vif(x):
    vif=pd.DataFrame()
    vif['variables']=x.columns
    vif['VIF']=[variance_inflation_factor(x.values,i)for i in range(x.shape[1])]
    return(vif)
```

```python
train_x,test_x,train_y,test_y=train_test_split(x1,y1,test_size=0.25)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(train_x,train_y)
log.score(train_x,train_y)
predlog=log.predict(test_x)
print(accuracy_score(test_y,predlog))
print(confusion_matrix(test_y,predlog))
print(classification_report(test_y,predlog))
```

```python

from sklearn.naive_bayes import MultinomialNB
```

```python
mnb=MultinomialNB()
mnb.fit(train_x,train_y)
mnb.score(train_x,train_y)
predmnb=mnb.predict(test_x)
print(accuracy_score(test_y,predmnb))
print(confusion_matrix(test_y,predmnb))
print(classification_report(test_y,predmnb))
```

as we can see that the accuracy_acore is far better than the logistic regression. lets check some more classification algorithms

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(train_x,train_y)
dtc.score(train_x,train_y)
preddtc=dtc.predict(test_x)
print(accuracy_score(test_y,preddtc))
print(confusion_matrix(test_y,preddtc))
print(classification_report(test_y,preddtc))
```

```python
from sklearn.svm import SVC
svc=svm.SVC(kernel='rbf')
svc.fit(train_x,train_y)
svc.score(train_x,train_y)
predsvc=svc.predict(test_x)
print(accuracy_score(test_y,predsvc))
print(confusion_matrix(test_y,predsvc))
print(classification_report(test_y,predsvc))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knc=KNeighborsClassifier()
knc.fit(train_x,train_y)
knc.score(train_x,train_y)
predknc=knc.predict(test_x)
print(accuracy_score(test_y,predknc))
print(confusion_matrix(test_y,predknc))
print(classification_report(test_y,predknc))
```

as we can se ethat decision tree classifier is workin best on my model, so lets find the best parameters using gridsearchcv

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
grid_param={'criterion':['gini','entropy'],'max_features':['auto','sqrt','log2']}
gd_sr=GridSearchCV(estimator=dtc,param_grid=grid_param,scoring='accuracy',cv=5)
gd_sr.fit(train_x,train_y)
best_parameters=gd_sr.best_params_
print(best_parameters)
best_result=gd_sr.best_score_
print(best_result)
```

```python
dtc=DecisionTreeClassifier(criterion='gini',max_features='sqrt')
dtc.fit(train_x,train_y)
dtc.score(train_x,train_y)
preddtc=dtc.predict(test_x)
print(accuracy_score(test_y,preddtc))
print(confusion_matrix(test_y,preddtc))
print(classification_report(test_y,preddtc))
```

lets try some ensemble techniques

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=10)
rfc.fit(train_x,train_y)
rfc.score(train_x,train_y)
predrfc=rfc.predict(test_x)
print(accuracy_score(test_y,predrfc))
print(confusion_matrix(test_y,predrfc))
print(classification_report(test_y,predrfc))
```
